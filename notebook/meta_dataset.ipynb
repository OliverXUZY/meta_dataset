## method 1

```python
s = 1; q = 9

source = 'aircraft'
deterministic = True
res = []
for idx in range(50):
    d = {}
    if deterministic:
        np.random.seed(idx)
    sampled_classes = np.random.choice(list(trainSet.class_images[source].keys()), 15, replace=False)
    d['source'] = source
    d['task_seed'] = idx
    tmp = {}
    for class_id in sampled_classes:
        # Get all available image indices for the class
        available_indices = trainSet.class_images[source][class_id]

        while len(available_indices) < s + q:  # If not enough images, resample class
            class_id = np.random.choice(list(trainSet.class_images[source].keys()))
            available_indices = trainSet.class_images[source][class_id]

        # Randomly sample images for support and query
        sampled_indices = np.random.choice(available_indices, s + q, replace=False)
        tmp[class_id] = sampled_indices.tolist()
    d['indices'] = tmp
    res.append(d)
```

## method 2

```python
s = 1; q = 9

source = 'aircraft'
deterministic = True
res = []
for idx in range(50):
    d = {}
    if deterministic:
        np.random.seed(idx)
    sampled_classes = np.random.choice(list(trainSet.class_images[source].keys()), 15, replace=False)
    d['source'] = source
    d['task_seed'] = idx
    flat_classes_images = []
    for class_id in sampled_classes:
        # Get all available image indices for the class
        available_indices = trainSet.class_images[source][class_id]

        while len(available_indices) < s + q:  # If not enough images, resample class
            class_id = np.random.choice(list(trainSet.class_images[source].keys()))
            available_indices = trainSet.class_images[source][class_id]

        # Randomly sample images for support and query
        sampled_indices = np.random.choice(available_indices, s + q, replace=False)
        flat_class_images = [(class_id, idx) for idx in sampled_indices]

        flat_classes_images += flat_class_images
    d['indices'] = flat_classes_images
    res.append(d)
```
